```python
import os
# 必须在导入transformers之前设置
os.environ["HF_ENDPOINT"] = "https://hf-mirror.com"

from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_name = "uer/gpt2-chinese-cluecorpussmall"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# 0：如果我拥有一台时间机器
# 1：当人类第一次踏上火星
# 2：如果动物会说话，它们最想告诉人类的是
# 3：有一天，城市突然停电了
# 4：当我醒来，发现自己变成了一本书
# 5：假如我能隐身一天，我会
# 6：我走进了那扇从未打开过的门
# 7：在一个没有网络的世界里
# 8：如果世界上只剩下我一个人
# 9：梦中醒来，一切都变了模样
#学号：202410179016
# 学号倒数第一位=6，对应开头
prompt = "我走进了那扇从未打开过的门"
# ====================================================

# 文本续写生成函数
def generate_text(prompt, max_length=200, temperature=0.7):
    # 编码输入文本
    inputs = tokenizer(prompt, return_tensors="pt")
    
    # 模型生成（避免重复、提升流畅度）
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_length=max_length,        # 生成总长度
            temperature=temperature,      # 生成随机性（越小越稳定）
            do_sample=True,              # 采样生成
            repetition_penalty=1.2,       # 重复惩罚
            pad_token_id=tokenizer.eos_token_id
        )
    
    # 解码并输出结果
    result = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return result

# 执行续写
if __name__ == "__main__":
    print("=" * 50)
    print(f"续写开头：{prompt}")
    print("=" * 50)
    generated = generate_text(prompt)
    print("生成结果：\n", generated)
```
